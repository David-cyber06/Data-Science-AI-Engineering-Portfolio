```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
```

```python
#Garantizar reproducibilidad
np.random.seed(42)
random.seed(42)
```

```python
# 1. Productos
categorias = ['Electrónicos', 'Ropa', 'Hogar', 'Alimentos', 'Deportes']
productos = []
for i in range(1, 51):
    categoria = random.choice(categorias)
    precio = random.randint(50, 500)
    productos.append({
        'producto_id': f'P{i:03d}',
        'nombre': f'Producto {i}',
        'categoria': categoria,
        'precio': precio,
        'costo': round(precio * random.uniform(0.4, 0.7), 2)
    })

df_productos = pd.DataFrame(productos)
```

```python
# 2. Clientes (200 clientes)
regiones = ['Bogotá', 'Medellín', 'Cali', 'Barranquilla', 'Bucaramanga']
clientes = []
for i in range(1, 201):
    clientes.append({
        'cliente_id': f'C{i:04d}',
        'region': random.choice(regiones)
    })

df_clientes = pd.DataFrame(clientes)
```

```python
# 3. Ventas (5000 transacciones)
ventas = []
start_date = datetime(2023, 1, 1)
end_date = datetime(2024, 12, 31)

for i in range(5000):
    cliente = random.choice(df_clientes['cliente_id'].tolist())
    producto = random.choice(df_productos['producto_id'].tolist())
    # Fecha aleatoria en el rango
    days = (end_date - start_date).days
    fecha = start_date + timedelta(days=random.randint(0, days))

    cantidad = random.randint(1, 5)
    producto_info = df_productos[df_productos['producto_id'] == producto].iloc[0]

    descuento = random.choice([0, 0.05, 0.10])
    total = producto_info['precio'] * cantidad * (1 - descuento)

    ventas.append({
        'venta_id': f'V{i+1:06d}',
        'cliente_id': cliente,
        'producto_id': producto,
        'fecha': fecha,
        'cantidad': cantidad,
        'precio_unitario': producto_info['precio'],
        'descuento': descuento,
        'total': round(total, 2)
    })

df_ventas = pd.DataFrame(ventas)
```

```python
# Guardar archivos
df_productos.to_csv('productos.csv', index=False)
df_clientes.to_csv('clientes.csv', index=False)
df_ventas.to_csv('ventas.csv', index=False)
```
